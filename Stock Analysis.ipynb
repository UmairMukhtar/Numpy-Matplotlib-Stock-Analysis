```python
import pandas as pd
```

```python
#Loading the excel file into data frame
df = pd.read_excel('Stock Analysis.xlsx')
```

```python
#Printing the content of the excel file
df
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 3

ind = np.arange(len(df['Total Revenue'][:4]))  # the x locations for the groups
width = 0.27       # the width of the bars

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)

yvals = [4, 9, 2]
rects1 = ax.bar(ind, df['Total Revenue'][:4], width, color='orange')
zvals = [1,2,3]
rects2 = ax.bar(ind+width, df['Net Income'][:4], width, color='gray')
kvals = [11,12,13]
rects3 = ax.bar(ind+width*2, df['Operating Cash Flow'][:4], width, color='b')

ax.set_ylabel('Revenue in Dollars')
ax.set_xticks(ind+width)
ax.set_xticklabels( ('2019', '2018', '2017','2016') )
ax.legend( (rects1[0], rects2[0], rects3[0]), ('Total Revenue', 'Net Income', 'Operating Cash Flow') )
        
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
N = 1
ind = np.arange(len(df['ROIC'][:4]))  # the x locations for the groups
width = 0.27       # the width of the bars

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)

yvals = [4, 9, 2]
rects1 = ax.bar(ind+0.3, df['ROIC'][:4], width, color='orange',label="ROIC")
ax.set_ylabel('Revenue in Percentage')
ax.set_xticks(ind+width)
ax.set_xticklabels( ('2019', '2018', '2017','2016') )
ax.legend()
plt.show()
```

```python
import matplotlib
N = 4
ind = np.arange(len(df['Total Cash'][:4]))  # the x locations for the groups
width = 0.21       # the width of the bars

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)


rects1 = ax.bar(ind, df['Total Cash'][:4], width, color='orange',label="Total Cash")

rects2 = ax.bar(ind+width, df['Total Current Assets'][:4], width, color='gray',label="Total Current Assets")

rects3 = ax.bar(ind+width+width, df['Total Assets'][:4], width, color='b',label='Total Assets')

rects4 = ax.bar(ind+width+width+width, df['Total Current Liabilities'][:4], width, color='yellow',label="Total Current Liabilities")
#ax.set_yticklabels(df['Total Assets'][:4], minor=False)
ax.yaxis.set_major_formatter(matplotlib.ticker.StrMethodFormatter('{x:,.0f}'))
ax.set_ylabel('Revenue in Dollars')
ax.set_xticks(ind+width)
ax.set_xticklabels( ('2019', '2018', '2017','2016') )
ax.legend()
        
plt.show()
```
